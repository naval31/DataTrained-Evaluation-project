# PREPROCESSING

```python
#dropping the column objid and rerun
df_space.drop(['objid','rerun'],axis=1,inplace=True)
```

### converting catergorical data to numerical data

```python
#using label encoder in attribute class
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_space['class']=le.fit_transform(df_space['class'])
```

```python
df_space
```

### Outliners

```python
# importing zscore
from scipy.stats import zscore
z_score=np.abs(zscore(df_space))
print(df_space.shape)
#removing all the values which are greater then threshold=3
df_x=df_space[(z_score<3).all(axis=1)]
# shape of new data after removing outliners
print(df_x.shape)
```

```python
df_x
```

```python
# split the data into x and y
x=df_space.drop(['class'],axis=1)
y=df_space['class']
```

```python
#converting all the data into one standard format by using MinMax scaler  
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
x=pd.DataFrame(sc.fit_transform(x),columns=x.columns)
```

```python
x
```

```python
# importing boxcox to remove skewness in the data
import matplotlib.pyplot as plt
from scipy.stats import boxcox

for col in x.columns:
    if x[col].skew()>0.55: 
        x[col]=np.log1p(x[col])
```

```python
#skewness after using boxcox
x.skew()
```

# Training Model

```python
#importing model
from sklearn.model_selection import train_test_split

#importing all the classification alogrithm 
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

#importing boosting technique
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier,BaggingClassifier,ExtraTreesClassifier

#importing metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
```

```python
#using classification algorithm with the help of loop
def model(m):
    max_r_score=0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=r_state,stratify=y)
        m.fit(x_train,y_train)
        predm=m.predict(x_test)
        a_score=accuracy_score(y_test,predm)
        if a_score>max_r_score:
            max_r_score=a_score
            final_r_state=r_state
    print("max accuracy score of ",m,"at random state",final_r_state,"is ",a_score)   
    
    print("confusion_matrix:\n  ",confusion_matrix(y_test,predm))
    print('classification_report\n  ',classification_report(y_test,predm))
    print("-------------------------------------------------------------------------------------------------------------")
```

```python
Model=[SVC(),DecisionTreeClassifier(),KNeighborsClassifier(),RandomForestClassifier(),AdaBoostClassifier(),BaggingClassifier(),ExtraTreesClassifier()]
for m in Model:
    model(m)
```

- __DecisionTreeClassifier,RandomForestClassifier,AdaBoostClassifier,BaggingClassifier__ all these alogrithm has accuracy approx 0.98

# Cross Validation

```python
#using the cross validation
from sklearn.model_selection import cross_val_score
Model=[SVC(),DecisionTreeClassifier(),KNeighborsClassifier(),RandomForestClassifier(),AdaBoostClassifier(),BaggingClassifier(),ExtraTreesClassifier()]
for i in Model:
    scores=cross_val_score(i,x,y,cv=10,scoring="accuracy")
    print("scores of ",i)
    print("mean score",scores.mean())
    print("Standard deviation",scores.std())
    print('\n-------------------------------------------------------------------------------------------------------')
```

- __DecisionTreeClassifier,RandomForestClassifier,BaggingClassifier__ all these alogrithm has __mean score approx 0.98__
- so we can choose anyone from these three as the both model giving accuracy score and mean score of 0.98 so there is no chance of overfitting so i am chossing __BaggingClassifier__ 

### Using GridSearchCV to get best parameters of BaggingClassifier

```python
#importing gridsearchcv
from sklearn.model_selection import GridSearchCV
dt=DecisionTreeClassifier()
bc = BaggingClassifier(dt,n_estimators = 100, max_features = 0.5)
param_grid = {
    'base_estimator__max_depth' : [1, 2, 3, 4, 5],
    'max_samples' : [0.05, 0.1, 0.2, 0.5]
}

bc_grid = GridSearchCV(estimator=bc,param_grid=param_grid, scoring='accuracy', cv=5)
bc_grid.fit(x,y)
best_params = bc_grid.best_params_
print(best_params)
```

# Best Model:BaggingClassifier

```python
#BaggingClassifier with best parameter
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=78,stratify=y)
dt=DecisionTreeClassifier(max_depth=3)
bc = BaggingClassifier(dt,n_estimators = 100, max_features = 0.5,max_samples=0.5)
bc=BaggingClassifier(n_estimators=40)
bc.fit(x_train,y_train)
bc_pred=bc.predict(x_test)
print("accuracy_score:",accuracy_score(y_test,bc_pred))
print("classification_report:\n",classification_report(y_test,bc_pred))
cm=confusion_matrix(y_test,bc_pred)
sns.heatmap(cm,annot=True,cbar=False)
plt.title("confusion matrix of bagging classifier")
```
