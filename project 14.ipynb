```python
#splliting interger value and removing string 
for col in range(0,12690):
    df_train.iloc[:,7][col] = df_train.iloc[:,7][col].replace("votes",'')
```

```python
df_train
```

```python
#converting first the rating New to nan values 
df_train["RATING"] = df_train["RATING"].replace("NEW",np.nan)
```

```python
#converting first the rating having '-' symbol  to nan values
df_train["RATING"] = df_train["RATING"].replace("-",np.nan)
```

```python
#converting the dataype of objevt to float
df_train["RATING"]=df_train["RATING"].astype(float)
```

```python
#filing the nan values with mean
df_train["RATING"]=df_train["RATING"].fillna(df_train["RATING"].mean())
```

```python
#converting the dataype of objevt to int
df_train["VOTES"]=df_train["VOTES"].astype(int)
```

```python
df_train.head()
```

```python
#dropping the unecessary columns
df_train.drop(["TITLE","RESTAURANT_ID","CUISINES","TIME"],axis=1,inplace=True)
```

```python
#filling city an locality with mode
df_train["CITY"] = df_train["CITY"].fillna(df_train["CITY"].mode()[0])
df_train["LOCALITY"] = df_train["LOCALITY"].fillna(df_train["LOCALITY"].mode()[0])
```

```python
#checking the null values
df_train.isnull().sum()
```

```python
#filling those null values with 0
df_train["c2"] = df_train["c2"].fillna(0)
df_train["c3"] = df_train["c3"].fillna(0)
df_train["c4"] = df_train["c4"].fillna(0)
df_train["c5"] = df_train["c5"].fillna(0)
df_train["c6"] = df_train["c6"].fillna(0)
df_train["c7"] = df_train["c7"].fillna(0)
df_train["c8"] = df_train["c8"].fillna(0)

```

```python
#now spliiting the cuisines column into 8 different columns 
df_te = pd.DataFrame(df_test.CUISINES.str.split(',',8).tolist(),
                                 columns = ['c1','c2','c3','c4','c5','c6','c7','c8'])

#now adding that cuisines dataset with the test data
df_test=pd.concat([df_test,df_te],axis=1)

#filling the null values in votes column
df_test["VOTES"]=df_test["VOTES"].fillna(df_test["VOTES"].mode()[0])

#splliting interger value and removing string 
for col in range(0,4231):
    df_test.iloc[:,7][col] = df_test.iloc[:,7][col].replace("votes",'')

#converting first the rating New to nan values 
df_test["RATING"] = df_test["RATING"].replace("NEW",np.nan)

#converting first the rating having '-' symbol  to nan values
df_test["RATING"] = df_test["RATING"].replace("-",np.nan)

#converting the dataype of objevt to float
df_test["RATING"]=df_test["RATING"].astype(float)

#filing the nan values with mean
df_test["RATING"]=df_test["RATING"].fillna(df_test["RATING"].mean())

#converting the dataype of objevt to int
df_test["VOTES"]=df_test["VOTES"].astype(int)

#dropping the unecessary columns
df_test.drop(["TITLE","RESTAURANT_ID","CUISINES","TIME"],axis=1,inplace=True)

#filling city an locality with mode
df_test["CITY"] = df_test["CITY"].fillna(df_test["CITY"].mode()[0])
df_test["LOCALITY"] = df_test["LOCALITY"].fillna(df_test["LOCALITY"].mode()[0])

#filling those null values with 0
df_test["c2"] = df_test["c2"].fillna(0)
df_test["c3"] = df_test["c3"].fillna(0)
df_test["c4"] = df_test["c4"].fillna(0)
df_test["c5"] = df_test["c5"].fillna(0)
df_test["c6"] = df_test["c6"].fillna(0)
df_test["c7"] = df_test["c7"].fillna(0)
df_test["c8"] = df_test["c8"].fillna(0)

#converting categorical data to numeric
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
l=["CITY","LOCALITY","c1","c2","c3","c4","c5","c6","c7","c8"]
for val in l:
    df_test[val]=le.fit_transform(df_test[val].astype(str))
```

# removing outliners 

```python
#import zscore
from scipy.stats import zscore
z=np.abs(zscore(df_train))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
# removing all values which are greater then thershold
df_new=df_train[(z<3).all(axis=1)]
```

# Feature selection¶

```python
df_new.shape
```

```python
df_new.columns
```

```python
#spillting data into input and output variable
x=df_new.drop('COST',axis=1)
x.head()
```

```python
y=df_new["COST"]
y.head()
```

```python
#checking the shape of input and output varaible
print(x.shape)
print(y.shape)
```

```python
#scaling the input variables
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
#spiltting the data into train and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=.30)
```

```python
#checking the shape
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#importing all the model libraries
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

#importing boosting models
from sklearn.ensemble import AdaBoostRegressor,RandomForestRegressor,GradientBoostingRegressor

#importing error metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.model_selection import GridSearchCV,cross_val_score
```

```python
model=[LinearRegression(),DecisionTreeRegressor(),KNeighborsRegressor(),SVR(kernel='poly'),Lasso(),Ridge(),ElasticNet(),AdaBoostRegressor(),RandomForestRegressor(),GradientBoostingRegressor()]
for m in model:    
    m.fit(x_train,y_train)
    print("score of ",m,"is :",m.score(x_train,y_train))
    pred=m.predict(x_test)
    print("\n")
    print("r2 score is:",r2_score(y_test,pred))
    print("mean absolute error is:",mean_absolute_error(y_test,pred))
    print("mean squared error is:",mean_squared_error(y_test,pred))
    print("Root mean squared error is:",np.sqrt(mean_squared_error(y_test,pred)))
    print("\n")
    print("                                                                  ")
    print("\n")
```

```python
#using the cross validation
from sklearn.model_selection import cross_val_score
model=[LinearRegression(),DecisionTreeRegressor(),KNeighborsRegressor(),SVR(),Lasso(),Ridge(),ElasticNet(),AdaBoostRegressor(),RandomForestRegressor(),GradientBoostingRegressor()]
for m in model:
    score=cross_val_score(m,x,y,cv=5,scoring='r2')
    print("score of ",m,"is:",score)
    print("mean score",score.mean())
    print("standard deviation",score.std())
    print('\n')
    print("           ")
    
```
