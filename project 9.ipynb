## In Recovered column we have seen some null values so we will fill these column using a simple formula
-                           __Recovered=Confirmed-Deaths-Active__       

```python
#filling the nan values of recovered column using formula
df_c['Recovered']=df_c['Confirmed']-df_c['Deaths']-df_c['Active']
```

```python
#check the data again and all the null values is replaced in recovered column
df_c.head()
```

- For State __American Samoa__ the number of __deaths__ is __Zero__,so __Mortality_rate__ will also be __zero__,so replacing __Nan__ value in __mortality_rate__ to __zero__

```python
#using fillna method to change null value to zero 
df_c['Mortality_Rate']=df_c['Mortality_Rate'].fillna(0)
```

### Dropping columns

```python
df_c.drop(["People_Hospitalized","Hospitalization_Rate","Lat","Long_","Incident_Rate"],axis=1,inplace=True)
```

__Reasons to drop above columns:__
   - "People_Hospitalized" column have nearly 45% of null values and secondly there is no another method to fill those even           using mean of column will be improper way of filling the data
   - Lat and Long are not of use for visualization

```python
df_c
```

```python
#checking again the null values 
df_c.isnull().sum()
```

```python
#Diamond Princess,Grand Princess needto be drop because people_tested column has nan value and also deaths are not much so we will not loose much data
df_c.drop([9,13],inplace=True)
```

```python
#reseting the index value
df_c.reset_index(drop=True,inplace=True)
```

```python
df_c
```

```python
#calculating the recovery_rate
df_c["Recovery_rate"]=df_c["Recovered"]*100/df_c["Confirmed"]
```

```python
#while calculationg recovering_rate we might end with some null values so instead of zero,so replacing null value to zero
df_c["Recovery_rate"]=df_c["Recovery_rate"].fillna(0)
```

# Some Calculations on the data to know the number of deaths,recovery,active case

```python
#calculating total mortality rate and recovery rate 
total_mortalityrate=df_c["Mortality_Rate"].sum()
total_recoveryrate=df_c["Recovery_rate"].sum()
```

```python
#output of total_mortalityrate
total_mortalityrate
```

```python
#output of total_recoveryrate
total_recoveryrate
```

```python
#total deaths in US
total_deaths=df_c["Deaths"].sum()
#total confirmed case in US
total_confirmed=df_c["Confirmed"].sum()
#total recovery in US
total_recovery=df_c["Recovered"].sum()
#total active case in US
total_active=df_c["Active"].sum()
```

### Making new DataFrame for above calculated data

```python
df_new=pd.DataFrame(data=[total_mortalityrate,total_recoveryrate,total_deaths,total_confirmed,total_recovery,total_active],index=['total_mortalityrate','total_recoveryrate','total_deaths','total_confirmed','total_recovery','total_active'])
```

```python
#Transpose the data
df_new=df_new.T
```

```python
#displaying new dataftrame after transpose
df_new
```
